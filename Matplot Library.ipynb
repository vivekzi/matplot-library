# Matplot 

```python
import matplotlib.pyplot as plot
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,100,101)
```

```python
x
```

```python
y=x**2
```

```python
y
```

## ploting



```python
plot.plot(x,y,'r-')
plot.xlabel('X')
plot.ylabel('Y')
```

```python
#OO

fig=plot.figure()
axes=fig.add_axes([0.1,0.1,.8,.8])

axes.set_xlabel('X-axis')
axes.set_ylabel('y-axis')
axes.plot(x,y)
```

```python
fige=plot.figure()
axe1=fige.add_axes([0.1,.1,.4,.6])
axe1.set_title('AAA')
axe=fige.add_axes([0.9,0.2,.8,.8])
axe.set_title('BBB')
```
