## ploting



```python
plot.plot(x,y,'r-')
plot.xlabel('X')
plot.ylabel('Y')
```

```python
#OO

fig=plot.figure()
axes=fig.add_axes([0.1,0.1,.8,.8])

axes.set_xlabel('X-axis')
axes.set_ylabel('y-axis')
axes.plot(x,y)
```

## Features

```python
fig=plot.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x*x,label='Square',color='red',lw='3',ls='steps')

axes.legend()
```

```python
fig=plot.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x*x,label='Square',color='red',lw='3',ls='-',marker='o',markersize='5')

axes.legend()
```
